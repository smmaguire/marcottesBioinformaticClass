```python
def freqDict(seq_filename):
    total_length = 0
    dic = {} # create an empty dictionary
    seq_file = open(seq_filename, "r")
    for raw_line in seq_file:
        line = raw_line.rstrip("\r\n")
        length = len(line)
        for nuc in line:
            if dic.has_key(nuc):
                dic[nuc] += 1
            else:
                dic[nuc] = 1
        total_length += length
    seq_file.close()
    newdic={}
    for n in dic.keys():
        fraction = float(dic[n]) / total_length
        newdic[n]=fraction
    return(newdic)
```

```python
# from hw1
#Same as question 2 just saves the frequencies in a dictionary
def digram(seq_filename):
    
    total_length = 0
    dinuc = {} # create an empty dictionary
    seq_file = open(seq_filename, "r")
    for raw_line in seq_file:
        line = raw_line.rstrip("\r\n")
        length = len(line)-1
        for count in range(length):
            di = line[count] + line[count+1]
            if dinuc.has_key(di):
                dinuc[di] += 1
            else:
                dinuc[di] = 1
        total_length += length+1
    seq_file.close()
    
    percentDi_ecoli= {}
    for n in dinuc.keys():
        fraction = dinuc[n] / float(total_length)
        percentDi_ecoli[n]=fraction
    return(percentDi_ecoli)
```

```python
def maxValDic(dic,what): #function to get the max vals in a dictionary or the key with the max val
    val=list(dic.values())
    key=list(dic.keys())
    if what=="key":
        return key[val.index(max(val))]
    else:
        return max(val)
```

```python
def getOther(state):
    if(state=="S"):
        return("T")
    elif(state=="T"):
        return("S")
```

```python
def recordMax(h,o,trans_p,before):
    a=before[h]*trans_p.ix[h,h]
    b=before[o]*trans_p.ix[o,h]
    if(a>b):
        return(h)
    if(b>a):
        return(o)
```

```python
def vitebri(obs, states,start_p,trans_p,emit_p):
    V=[{}] #list of dictionaries that will build the path matrix
    paths={}
    paths[1]=[]
    paths[2]=[]
    
    for hidden in states: #initialize
        V[0][hidden] = start_p[hidden]*emit_p[hidden][obs[0]]
        paths[1].append(maxValDic(V[0],"key"))
        paths[2].append(maxValDic(V[0],"key"))
    for t in range(1,len(obs)): #itterate through a range the length of the observations starting at 1
        given=obs[t] #the amino acid observed at a particular site
        before=V[t-1] #the dictionary containing the probabilities from the last round
        V.append({})#Add an empty dictionary to the list
        testMax={}
        for hidden in states:
            other=getOther(hidden)
            emitObs=emit_p[hidden][obs[t]]
            testMax[hidden]=max((before[hidden]*trans_p.ix[hidden,hidden]),(before[other]*trans_p.ix[other,hidden]))
            if(hidden=="S"):
                paths[1].append(recordMax(hidden,other,trans_p,before))
            elif(hidden=="T"):
                paths[2].append(recordMax(hidden,other,trans_p,before))
            V[t][hidden]=testMax[hidden]*emitObs
    if(V[t]["S"]>V[t]["T"]):
        print(paths[1])
    else:
        print(paths[2])
    return(V)
```

###2. What are the relative proportions of transmembrane and non-transmembrane amino acid segments (i.e, segments of all T or all S) in integral membrane proteins?

```python
seq_file = open("state_sequences.txt", "r")
Svector=[]
Tvector=[]
for raw_line in seq_file:
    line = raw_line.rstrip("\r\n")
    length = len(line)-1
    counterS=0
    counterT=0
    for count in range(length):
        current=line[count]
        next=line[count+1]
        if(current!=next and current=="S"):
            counterS+=1
        elif(current!=next and current=="T"):
            counterT+=1
        elif(count==length-1 and next =="S"):
            counterS+=1
        elif(count+1==length-1 and next =="T"):
            counterT+=1
    Tvector.append(counterT)
    Svector.append(counterS)

proportion=[]
for count2 in range(len(Tvector)):
    #print(count2)
    if(Svector[count2]!=0):
        proportion.append(Tvector[count2]/float(Svector[count2]))
    else:
        proportion.append(1)
```

```python
import matplotlib.pyplot as plt
import numpy as np

hist, bins = np.histogram(proportion,bins = 20)
width = 0.7*(bins[1]-bins[0])
center = (bins[:-1]+bins[1:])/2
plt.bar(center, hist, align = 'center', width = width)
plt.show()
```
