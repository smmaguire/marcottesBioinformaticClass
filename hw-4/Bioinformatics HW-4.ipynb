### Loads the entire file into a dictionary with the protiens appeneded into one full line and the names as the keys

```python
import time
#gets file into dictionary with all the protiens appended
seqList = list(line.rstrip("\n").split(None,1)[0] for line in open("yeast_aaseqs.txt"))
proData=dict()
for i in range(len(seqList)):
    if seqList[i][0]=='>':
        fullPro = []
        for j in range((i+1),len(seqList)):
            if seqList[j][0]=='>':
                if seqList[i] in proData:
                    name=str(time.clock())+seqList[i] # make sure they have unique names
                else:
                    name = seqList[i]
                proData[name]=''.join(fullPro)
                break
            else:
                fullPro.append(seqList[j])
```

### calculate the AA for each protien in the dictionary, create a new dictionary for each with those frequencies make a list of all the dictionaries, make the list into a pandas dataframe

```python
import re
import pandas as pd
#make a new dictionary with the frequency of each amino acid 
freqList=[]# list of dictionaries of amino acid frequencies
indexList=[] # keep track of the names
for keys in proData:
    indexList.append(keys)
    protein=proData[keys]
    newDic={'G':0,'A':0,'V':0,'L':0,'I':0,'P':0,'F':0,'Y':0,'W':0,'S':0,'T':0,'C':0,'M':0,'N':0,'Q':0,'K':0,'R':0,'H':0,'D':0,'E':0}
    for AA in newDic:
        newDic[AA] = (len(re.findall(AA,protein))/float(len(protein)))
    freqList.append(newDic)
df1=pd.DataFrame(freqList,index=indexList)
```

### turn the pd.data.frame into a tab delimted text file

```python
df1.to_csv('protien.csv',sep='\t')
```

### Heirarchical clustering cut at a certain point to only show a reasonable number of groups for plotting, in this case about 17 clusters are shown

```python
%%R
data=read.table("protien.csv",sep='\t',header=TRUE)
names(data)[1]<-'pro_name'
# Ward Hierarchical Clustering
d <- dist(data[,2:ncol(data)], method = "euclidean") # distance matrix
fit <- hclust(d, method="ward")
#plot(fit,labels=data[,1]) # display dendogram
dendro<-as.dendrogram(fit)
plot(cut(dendro, h=3)$upper,)
clust<-cutree(fit,h=3)
```
