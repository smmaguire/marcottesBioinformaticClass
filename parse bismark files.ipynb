## This a script to parse CpG hits from bismark and add the results to one large table

```python
#helper function to pull the sample name out of the file name
import re
def getName(fileName):
    new=[]
    pn= re.search( r'(f|m)(.*)', fileName, re.M|re.I)
    if pn.group()[2]=='_':
        return (pn.group()[0:2])
    else:
        return (pn.group()[0:3])
```

```python
import pandas as pd
cpgList= pd.read_csv("CpG_list.csv")
cpgList=cpgList.set_index('site')
files =!ls CpG*.txt
fullDat=pd.DataFrame()
for nextFile in files:
    data = pd.read_table(nextFile, header=None, skiprows=1, sep= r'\s*')
    data=data.ix[:,1:]
    data=data.rename(columns={'X.2':'meth1','X.3':'chrom','X.4':'site','X.5':'meth2'})
    grouped = data.groupby('site')
    test2=grouped.count()
    grouped2 =data.groupby(['meth2','site'])
    test3= grouped2.count()
    test4=test3.xs('Z',level='meth2')
    test5=pd.DataFrame(test4['meth1'])
    test6=pd.DataFrame(test2['chrom'])
    test7=test6.join(test5)
    test7['meth1']=test7['meth1'].fillna(0)
    test7['per_methyl']=test7['meth1'].astype(float)/(test7['chrom'])
    test7=cpgList.join(test7)
    test7['sample']=getName(nextFile)
    test7=test7.rename(columns={'chrom':'coverage'})
    test7=test7.reset_index()
    fullDat=pd.concat([fullDat,test7])
```

## Plot Methylation percentages at each cpg site

```python
from matplotlib.backends.backend_pdf import PdfPages
colors=['y','b','g']
colorCount=0
pdf =PdfPages('methylation% by sample.pdf')
for key, grp in fullDat.groupby(['sample']):
    plt.figure()
    x=(grp['site'])
    y=(grp['per_methyl'])
    plt.bar(x,y,.1,alpha=.4,label=key,color='g',edgecolor='g')
    colorCount += 1
    plt.ylim([-.2,1.2])
    plt.xlim([2900,6000])
    plt.legend(loc='upper left')
    plt.xlabel('CpG site location')
    plt.ylabel('methylation percentage')
    if str(key)!= "f17" and str(key)!= "m28":
        pdf.savefig()
    plt.show()
pdf.close()
```
